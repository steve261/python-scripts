## FIGURES AND SUB PLOTS

```python
import numpy as np
import matplotlib.pyplot as plt
x2=np.linspace(1,1000,10)
y3=np.log(x2)
y4=np.cos(x2)
## create a figure with two sub plots(1 row, 1 column)

fig, axs=plt.subplots(1,2, figsize=(10, 4))
##1D -array
## the first subplot at index 0
axs[0].plot(x2,y3, label="Log(x)",color ='blue')
axs[0].set_title("SUBPLOT")
axs[0].set_xlabel("x-axis")
axs[0].set_ylabel("y-axis")
axs[0].legend()
                   
##THE SECOND PLOT
axs[1].plot(x2, y4, label='Cos(x)', color='red')
axs[1].set_title('Cosine Function')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```

```python
fig, axs=plt.subplots(2,2)
xx=np.linspace(1,1000,10)
yy1=np.log(x2)
yy=np.cos(x2)
## first sub plot
axs[0,0].plot([1,2,3,4],[2,4,7,8])
axs[0,0].set_title("SUB PLOT 1")
axs[0,0].set_xlabel("x-axis")
axs[0,0].set_ylabel("y-axis")

## second sub plot
axs[0,1].plot(xx,yy)
axs[0,1].set_title("SUB PLOT 2")
axs[0,1].set_xlabel("x-axis")
axs[0,1].set_ylabel("y-axis")

# Third subplot
axs[1,0].plot(xx, yy1)
axs[1,0].set_title('Subplot 3')
axs[1,0].set_xlabel("x-axis")
axs[1,0].set_ylabel("y-axis")

# Fourth subplot
axs[1,1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axs[1,1].set_title('Subplot 4')
axs[1,1].set_xlabel("x-axis")
axs[1,1].set_ylabel("y-axis")

## Adjust layout to prevent overlap
plt.tight_layout()

##Show the figure
plt.show()
```
